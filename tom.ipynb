# Pandas - Data Covariance


Covariance shows the relationship between two variables.

It tells whether the variables increase or decrease together (positive covariance) or move in opposite directions (negative covariance).

##### 1.Positive Covariance:
If one variable increases and the other also increases, the covariance is positive.

##### 2.Negative Covariance:
If one variable increases and the other decreases, the covariance is negative.

##### 3.Zero Covariance:
If the variables are not related, the covariance will be close to zero.


```python

print(project.cov(numeric_only=True))
```

## Questions
1.What is the total sales revenue generated by each branch ?

```python
total_sales_by_branch =project.groupby('Branch')['Total'].sum()

print(total_sales_by_branch)


```

2.Which product line has the highest total sales revenue?

```python

total = project.groupby('Product_line')['Total'].sum().sort_values(ascending=False)

total.head(2)
```

3.What is the average total sales revenue per transaction?

```python
average_sales_per_transaction = project['Total'].mean()
print("The average is:",average_sales_per_transaction)
```

4.What is the sales distribution between customer types (Member vs. Normal)?

```python
sales_distribution_by_customer_type = project.groupby('Customer_type')['Total'].sum()
print(sales_distribution_by_customer_type)
```

5.How do male and female customers compare in terms of average spending?

```python
average_spending_by_gender = project.groupby('Gender')['Total'].mean()


print(average_spending_by_gender)
```

6.What is the average customer rating for each branch?

```python
average_rating_by_branch = project.groupby('Branch')['Rating'].mean()

# Display the result
print(average_rating_by_branch)
```

7.Which product line has the highest average unit price?

```python
product_line = project.groupby('Product_line')['Unit_price'].mean()

highest_avg_unit_price = product_line.sort_values(ascending=False).head(1)

print(highest_avg_unit_price)
```

8.What is the average quantity sold for each product line?


```python
a=project.groupby('Product_line')['Quantity'].mean()
print(a)
```

9.How does gross income vary across different product lines?

```python
gross_income_by_product_line = project.groupby('Product_line')['Gross_income'].sum()

plt.figure(figsize=(10, 6))
gross_income_by_product_line.plot(kind='bar', color='skyblue')
plt.title('Gross Income Across Different Product Lines',color='red')
plt.xlabel('Product Line',color='red')
plt.ylabel('Total Gross Income',color='red')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(gross_income_by_product_line)
```

10.Which city generates the highest sales revenue?

```python
sales_by_city = project.groupby('City')['Total'].sum()

highest_sales_city = sales_by_city.sort_values(ascending=False).head(1)

print(highest_sales_city)
```

11.What is the most frequently used payment method?



```python
payment_method_counts = project['Payment'].value_counts()

most_frequent_payment_method = payment_method_counts.head(1)
print(most_frequent_payment_method)
```

12.What is the average customer rating for each city?


```python
average_rating_by_city = project.groupby('City')['Rating'].mean()

print(average_rating_by_city)
```

13.What is the average tax collected for each product line?



```python
tax=project.groupby('Product_line')['Tax_5%'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
tax.plot(kind='bar', color='skyblue')
plt.title('Average Tax Collected for Each Product Line',color='red')
plt.ylabel('Average Tax',color='red')
plt.xlabel('Product Line',color='red')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


print(tax)
```

14.What is the sales trend by month?


```python
project['Month'] = project['Date'].dt.to_period('M')

sales_trend_by_month = project.groupby('Month')['Total'].sum()

plt.figure(figsize=(10, 6))
sales_trend_by_month.plot(kind='line', marker='o', color='k', linestyle='-', linewidth=2, markersize=6)
plt.title('Sales Trend by Month',color='red')
plt.xlabel('Month',color='red')
plt.ylabel('Total Sales Revenue',color='red')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

15.Which branch has the highest average customer rating?



```python
average_rating_by_branch = project.groupby('Branch')['Rating'].mean()

highest_rating_branch = average_rating_by_branch.idxmax()

print(f"The branch with the highest average customer rating is:", highest_rating_branch)
```
