### Introduction

# Pandas - Data Covariance


Covariance shows the relationship between two variables.

It tells whether the variables increase or decrease together (positive covariance) or move in opposite directions (negative covariance).

##### 1.Positive Covariance:
If one variable increases and the other also increases, the covariance is positive.

##### 2.Negative Covariance:
If one variable increases and the other decreases, the covariance is negative.

##### 3.Zero Covariance:
If the variables are not related, the covariance will be close to zero.


```python
df.cov(numeric_only=True)

```

# Questions
1.What is the average overall score for universities in each country?

```python
avg = df.groupby('Country')['Overall_Score'].mean()
print(avg)
```

2.What is the distribution of faculty-student ratios across the dataset?

```python
faculty_student_ratio_distribution = df['Faculty_Student_Ratio'].describe()
print(faculty_student_ratio_distribution)
```

3.Which countries have the most universities in the top 100?

```python
top_100_countries = df.nlargest(50, 'Overall_Score')['Country'].value_counts()
print(top_100_countries)
```

4.What are the top 10 universities based on the overall score?

```python
top_10_universities = df.nlargest(10, 'Overall_Score')[[ 'University_Name', 'Overall_Score']]
print(top_10_universities)

```

5.What are the top 5 universities in terms of international faculty?

```python
top_2_international_faculty = df.nlargest(5, 'International_Faculty')[['University_Name', 'International_Faculty']]
print(top_2_international_faculty)
```

6.Which universities have p scors (90) in any of the categories?


```python
perfect_scores = df[(df == 90).any(axis=1)][['University_Name']]
print(perfect_scores)
```

7.What is the average employer reputation score for universities in China?

```python
avg_employer_reputation_china = df[df['Country'] == 'China']['Employer_Reputation'].mean()
print('avgerage=',avg_employer_reputation_china)
```

8.What is the standard deviation of overall scores in the dataset?

```python
stdd = df['Overall_Score'].std()
print(stdd)
```

9.How many universities have employer reputation scores above 95?

```python
a = df[df['Employer_Reputation'] > 95].shape[0]
print("employer rep above 95 are:",a)
```

10.Which universities have the lowest scores in Overall Score?

```python
lowest = df.nsmallest(5, 'Overall_Score')[['University_Name', 'Overall_Score']]
print(lowest)
```

11.Are there universities with the same overall score but different ranks?

```python
duplicate_scores = df[df.duplicated('Overall_Score', keep=False)][['Rank', 'University_Name', 'Overall_Score']]
print(duplicate_scores)
```

12.What is the overall ranking trend for universities located in Singapore?

```python
singapore_universities = df[df['Country'] == 'Singapore'][['Rank', 'University_Name']]
print(singapore_universities)
```

13.How does the faculty-student ratio affect the overall score?

```python
correlation = df['Faculty_Student_Ratio'].corr(df['Overall_Score'])
print(correlation)
```

14.Which universities have the best scores in citations per paper and papers per faculty combined?

```python
df['Combined_Score'] = df['Citations_per_Paper'] + df['Papers_per_Faculty']
best_combined_scores = df.nlargest(5, 'Combined_Score')[['University_Name', 'Combined_Score']]
print(best_combined_scores)
```

15.How does the rank correlate with the international student percentage?

```python
correlation = df['Rank'].corr(df['International_Students'])
print(correlation)
```
